Scatter plot: Visualize the relationship between 'orbital_period' and 'mass' of the planets.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Planets dataset from seaborn
planets = sns.load_dataset('planets')

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='orbital_period', y='mass', data=planets)
plt.title('Relationship between Orbital Period and Mass of Planets')
plt.xlabel('Orbital Period (days)')
plt.ylabel('Mass (Jupiter Masses)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Bar plot: Display the count of planets discovered by each method

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Planets dataset from seaborn
planets = sns.load_dataset('planets')

# Count the number of planets discovered by each method
method_counts = planets['method'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=method_counts.index, y=method_counts.values, palette='viridis')
plt.title('Number of Planets Discovered by Each Method')
plt.xlabel('Discovery Method')
plt.ylabel('Number of Planets')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

Histogram: Visualize the distribution of 'distance' of planets from their respective stars.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Planets dataset from seaborn
planets = sns.load_dataset('planets')

# Filter out NaN values in the 'distance' column
distance_data = planets['distance'].dropna()

# Plotting
plt.figure(figsize=(10, 6))
sns.histplot(distance_data, bins=30, kde=True, color='skyblue')
plt.title('Distribution of Distance of Planets from Their Stars')
plt.xlabel('Distance (parsecs)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Pair plot: Show pairwise relationships between 'orbital_period', 'mass', 'year', and 'distance'.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Planets dataset from seaborn
planets = sns.load_dataset('planets')

# Select columns of interest
columns_of_interest = ['orbital_period', 'mass', 'year', 'distance']

# Create pair plot
sns.pairplot(planets[columns_of_interest].dropna(), diag_kind='hist')
plt.suptitle('Pairwise Relationships between Orbital Period, Mass, Year, and Distance')
plt.tight_layout()
plt.show()
```

Violin plot: Compare the distribution of 'year' for different 'method' of planet discovery.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Planets dataset from seaborn
planets = sns.load_dataset('planets')

# Filter out NaN values in the 'year' column
planets_filtered = planets.dropna(subset=['year'])

# Plotting
plt.figure(figsize=(12, 8))
sns.violinplot(x='method', y='year', data=planets_filtered, palette='Set2')
plt.title('Distribution of Discovery Year by Method of Planet Discovery')
plt.xlabel('Method of Discovery')
plt.ylabel('Discovery Year')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Swarm plot: Visualize the 'mass' of planets discovered by different 'method'


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Planets dataset from seaborn
planets = sns.load_dataset('planets')

# Filter out NaN values in the 'mass' column
planets_filtered = planets.dropna(subset=['mass'])

# Plotting
plt.figure(figsize=(12, 8))
sns.swarmplot(x='method', y='mass', data=planets_filtered, palette='viridis')
plt.title('Mass of Planets Discovered by Method')
plt.xlabel('Method of Discovery')
plt.ylabel('Mass (Jupiter Masses)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Heatmap: Create a heatmap to show the correlation matrix between numerical columns in the dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Planets dataset from seaborn
planets = sns.load_dataset('planets')

# Select numerical columns for correlation analysis
numerical_columns = ['number', 'orbital_period', 'mass', 'distance', 'year']

# Compute the correlation matrix
correlation_matrix = planets[numerical_columns].corr()

# Plotting
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numerical Columns')
plt.tight_layout()
plt.show()
```

Point plot: Compare the mean 'orbital_period' for each 'year' of planet discovery

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Planets dataset from seaborn
planets = sns.load_dataset('planets')

# Filter out NaN values in the 'orbital_period' column
planets_filtered = planets.dropna(subset=['orbital_period'])

# Plotting
plt.figure(figsize=(12, 8))
sns.pointplot(x='year', y='orbital_period', data=planets_filtered, ci='sd', capsize=0.2, color='skyblue')
plt.title('Mean Orbital Period by Year of Discovery')
plt.xlabel('Year of Discovery')
plt.ylabel('Mean Orbital Period (days)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Bar plot with error bars: Show the mean 'mass' of planets for different 'method' of discovery with confidence intervals.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the Planets dataset from seaborn
planets = sns.load_dataset('planets')

# Filter out NaN values in the 'mass' column
planets_filtered = planets.dropna(subset=['mass'])

# Calculate mean and confidence intervals
mean_mass = planets_filtered.groupby('method')['mass'].mean()
ci_mass = planets_filtered.groupby('method')['mass'].std() / np.sqrt(planets_filtered.groupby('method')['mass'].count()) * 1.96

# Plotting
plt.figure(figsize=(10, 6))
mean_mass.plot(kind='bar', yerr=ci_mass, capsize=5, color='skyblue')
plt.title('Mean Mass of Planets by Discovery Method')
plt.xlabel('Method of Discovery')
plt.ylabel('Mean Mass (Jupiter Masses)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
